```
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation, Flatten,Conv2D, MaxPooling2D
from keras.callbacks import TensorBoard
import pickle
import time


X=pickle.load(open("X.pickle","rb"))
y=pickle.load(open("y.pickle","rb"))

X=X/255.0
i=0

NAME="papers-{}".format(  int(time.time())  )
tensorboard= TensorBoard(log_dir='./individual/{}'.format(NAME))

model=Sequential()

model.add(   Conv2D(20, (5,5)   , input_shape= X.shape[1:]) )
model.add(   Activation("relu") )
model.add(   MaxPooling2D(pool_size=(4,4)))

model.add(   Conv2D(50, (5,5)  ) )
model.add(   Activation("relu") )
model.add(   MaxPooling2D(pool_size=(6,6)))
#model.add( Dropout(0.1) )

model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))
#model.add(Dropout(0.1))

model.add(Dense(1))
model.add(Activation('sigmoid'))


model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
model.fit(X,y,  epochs=20, batch_size=64 , verbose=1 , validation_split=0.15,  callbacks=[tensorboard])
print("\n\n\n")

model.save('CNN.model')
```
